```python
import os
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn import svm
from sklearn.metrics import accuracy_score
import joblib
print("Libraries imported")
```

```python
parkinson_data = pd.read_csv("parkinsons.csv")
parkinson_data.head(10) #printing first 10 rows of dataframe
```

```python
parkinson_data.shape #rows and columns in dataset
```

```python
parkinson_data.describe() #statistical data about the dataset
```

```python
parkinson_data.isnull().sum() #checking for missing values
```

```python
parkinson_data.info() #getting more info about the dataset
```

```python
X = parkinson_data.drop(columns=['status', 'name'], axis=1)
```

```python
# Exclude non-numeric columns from the correlation calculation
numeric_columns = parkinson_data.select_dtypes(include=[np.number]).columns
correlation_matrix = parkinson_data[numeric_columns].corr()

# Plot the correlation matrix
plt.figure(figsize=(12, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Matrix')
plt.show()
```

```python
# target variable = status 
parkinson_data['status'].value_counts()
```

```python
non_numeric_columns = parkinson_data.select_dtypes(exclude=[np.number]).columns
print(non_numeric_columns)
```

```python
non_numeric_columns = parkinson_data.select_dtypes(exclude=[np.number]).columns
parkinson_data.drop(columns=non_numeric_columns, inplace=True)
```

```python
parkinson_data.groupby('status').mean()
```

```python
Y = parkinson_data['status']
```

```python
print(Y)
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=2)
```

```python
print(X.shape, X_train.shape, X_test.shape)
```

```python
scaler = StandardScaler()
```

```python
scaler.fit(X_train)
```

```python
# Save the scaler
joblib.dump(scaler,'skaler.pkl')
```

```python
X_train = scaler.transform(X_train)
X_test = scaler.transform(X_test)
```

```python
print(X_train)
```

```python
model = svm.SVC(kernel='linear')
```

```python

model.fit(X_train, Y_train)
```

```python
# Accuracy Score on training data
X_train_pred = model.predict(X_train)
training_data_accuracy = accuracy_score(Y_train, X_train_pred)
```

```python
print('Accuracy (Training Data) :', training_data_accuracy*100, '%')
```

```python
# Accuracy Score on test data
X_test_pred = model.predict(X_test)
testing_data_accuracy = accuracy_score(Y_test, X_test_pred)
```

```python
print('Accuracy (Testing Data) :', testing_data_accuracy*100, '%')
```
